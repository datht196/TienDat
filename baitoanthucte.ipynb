Cho file"chipotle.tsv" về dữ liệu bán hàng của 1 nhà hàng.
Nhà hàng muốn lên danh sách thống kê xem mặt hàng nào bán được nhiều nhất,ít nhất để có chiến lược kinh doanh hợp lý.


```python
import pandas as pd
```

```python
df=pd.read_csv("chipotle.tsv",sep="\t")
df
```

```python
#B1:dùng hàm group để phân loại mỗi loại hàng
#B2;Dùng hàm sum để tính tổng số lượng bán được mỗi loại hàng
#B3:Dùng hàm sort để sắp xếp.
group_item=df.groupby("item_name")
sum_item=group_item["quantity"].sum()
sum_item.sort_values()
```

Vậy Carnitas Salad là mặt hàng bán được ít nhất và Chicken Bowl là mặt hàng bán được nhiều nhất
