```python
#import thư viện
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

sns.set_theme() #Apply the default theme
```

```python
#Các data set mà seaborn cung cấp
print(sns.get_dataset_names())
```

```python
#load data frame
df = sns.load_dataset('taxis')
df
#Dữ liệu về các chuyến taxi
```

## DISTRIBUTION PLOT

Phân phối đơn biến của các quan sát

```python
sns.histplot(data=df["total"]);
# Phân phối của column "total"
```

```python
df["total"].value_counts().sort_values(ascending=False)
```

Giống như trên đồ thị,ta có thể nhận thấy total(Chi phí của chuyến đi)thường tập trung trong khoảng(9.30,10.30)

## KDE plot
biểu đồ ước tính mật độ hạt nhân

KDE biểu diễn dữ liệu bằng cách sử dụng đường cong mật độ xác suất liên tục trong một hoặc nhiều chiều

```python
sns.kdeplot(data=df['total']);
```

## DIS plot

```python
sns.displot(data=df, x="total", col="payment", kde=True);
```

## Bar plot

```python
sns.barplot(data=df, x="passengers", y="tip", estimator=np.mean);
```

## Count plot

```python
sns.countplot(data=df, x="passengers");
```

## Box plot
Đồ thị là biểu đồ diễn tả 5 vị trí phân bố của dữ liệu, đó là: giá trị nhỏ nhất (min), tứ phân vị thứ nhất (Q1), trung vị (median), tứ phân vị thứ 3 (Q3) và giá trị lớn nhất (max).

Box plot có tác dụng:

Xác định ngoại lệ 

Kiểm tra xem dữ liệu có đối xứng không

Xác định mức độ chặt chẽ của dữ liệu được nhóm lại

```python
df.head()
```

```python
sns.boxplot(data=df, x="passengers", y="total",hue="payment",  palette='Blues');

plt.legend(loc=0);
```

## Facet Grid
Đồ thị vẽ tối đa 3 chiều:row,col,hue

```python
fg = sns.FacetGrid(df,  col="color",hue='payment') #Create a class instance of Facet Grid class
fg.map(sns.scatterplot,  'total','tip').add_legend();
#B
```

## JOINT PLOT
Đồ thị thể hiện sự tương quan giữa 2 biến khác nhau

```python
sns.jointplot(df, x="distance", y="total");
```

## PAIRS PLOT
Để vẽ nhiều phân phối hai biến theo cặp trong một tập dữ liệu, bạn có thể sử dụng

Các biểu đồ ở đường chéo là biểu đồ đơn biến

```python
sns.pairplot(df, hue="payment")
```
