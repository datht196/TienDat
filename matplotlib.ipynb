# Pyplot API vs Object-Oriented (OO) API
Pyplot API → Quickly

OO API → Advanced


# Pyplot API

```python
x = np.linspace(0, 10, 1000) #Sinh 1000 số từ 0 đến 10 với khoảng cachs giữa 2 số liên tiếp là như nhau
plt.plot(x, np.sin(x), color="blue", linestyle="dashed", label="sin(x)");
plt.plot(x, np.cos(x), color="red", label="cos(x)")

plt.title("Simple plot")#title của đồ thị
plt.xlabel("X");#tên của trục x
plt.ylabel("F(x)");#tên của trục y

plt.legend();#cách nhận biết đâu là sin(x) đâu là cos(x)
```

# Object-Oriented API

```python
x=np.linspace(0,2,1000)
fig, ax = plt.subplots(figsize=(20,20))
ax.plot(x, np.sin(x), label='sin(x)')  # Plot some data on te axes.
ax.plot(x, np.cos(x), label='cos(x)')  # Plot more data on the axes...
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

```python
#line
#OO API
x = np.linspace(0,10, 100)
x[:5]
fig, ax = plt.subplots()
ax.plot(x, x**3);
ax.set_xlabel('x')  # Add an x-label to the axes.
ax.set_ylabel('x^3')  # Add a y-label to the axes.
ax.set_title("Simple Plot")
```

```python
#PYPLOT API
x = np.linspace(0,10, 100)
plt.plot(x,x**3)
plt.title("Simple Plot")#title của đồ thị
plt.xlabel("x");#tên của trục x
plt.ylabel("x^3");#tên của trục y
```

```python
#scatter
#pyplot api
plt.scatter(x, np.exp(x)); #y=e^(x)
```

```python
#Bar
soft_drink_prices = {"Coke": 10,
                     "Pessi": 12,
                     "Sprite": 8}
dict(soft_drink_prices)
fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax.set(title="Bach Hoa Xanh's Soft Drink Prices", ylabel="Price ($)");
```

```python
#Histogram
np.random.seed(42)
#Tạo 1 mảng gồm các giá trị ngẫu nhiên tuân theo phân phối chuẩn với mean và std đã biết
student_height = np.random.normal(170, 10, 250)
plt.hist(student_height, bins=30);
```

```python
#subplots
#tạo figure có 2x2 axes
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
#thêm dữ liệu vào các axes
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax4.hist(student_height, bins=30);
```

Có 93 vị trí công việc khác nhau trong ngành khoa học dữ liệu.Tiếp theo chúng ta sẽ tìm top 10 công việc có nhiều người làm nhất trong năm 2023

```python

top10 = df[df["work_year"]==2023]['job_title'].value_counts()[:10].reset_index()
top10
```

```python

fig, ax = plt.subplots(figsize=(10, 6))
plt.style.available
plt.style.use('seaborn-v0_8-whitegrid')
ax.barh(top10['index'],top10['job_title']);
ax.set(ylabel='Job Titles',xlabel='Counts', title='Top 10 Job Titles in 2023')
```
